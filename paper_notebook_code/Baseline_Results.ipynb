# Prospectivity heatmaps of two baselines, i.e., sharing no data (the no-data-sharing case) and sharing all data (the full-data-sharing case).

### Figure 6 within the paper

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys

# Import necessary modules from the local 'minerals' package
sys.path.append('..')
from minerals import *

# Set the side length value
side_len = 50

# Load the final minerals grid data from a CSV file
df_minerals_grid_final = pd.read_csv('../data/baseline_option_' + str(side_len) + '.csv')

# Create subplots with 1 row and 2 columns, each with a specific size
f, ax = plt.subplots(1, 2, figsize=(15 * 2, 13))

# Adding new columns to the DataFrame
df_minerals_grid_final['abs_error_orig'] = 0
df_minerals_grid_final['proba_prospectivity_orig'] = df_minerals_grid_final['proba_prospectivity-40']

# Display the graph using the display_graph function
# Parameters used: ax[0] is the axis, df_minerals_grid_final is the DataFrame,
# 'orig' is a label, 'No data shared' is a title, lls=[5] is a list of labels,
# and conf_loss_count=0 is a configuration loss count
display_graph(ax[0], df_minerals_grid_final, 'orig', 'No data shared', lls=[5], conf_loss_count=0)

# Load another set of final minerals grid data from a CSV file
df_minerals_grid_final2 = pd.read_csv('../data/df_minerals_grid_' + str(side_len) + '_v2.csv')

# Adding new columns to the second DataFrame
df_minerals_grid_final2['abs_error_orig'] = 0
df_minerals_grid_final2['proba_prospectivity_orig'] = df_minerals_grid_final2['proba_prospectivity-40']

# Display the graph using the display_graph function for the second set of data
display_graph(ax[1], df_minerals_grid_final2, 'orig', 'All data shared')

# Show the plots
plt.show()

# Create a new subplot
f, ax = plt.subplots(1, 1, figsize=(13, 10))

# Display 'No data shared'
print('No data shared')
# Display the graph using the display_graph function and save the image
display(display_graph(ax, df_minerals_grid_final, 'orig', '', lls=[5], conf_loss_count=0))
filename = 'size_' + str(side_len) + '_nodatashared.png'

# Show the plot
plt.show()

# Create another subplot
f, ax = plt.subplots(1, 1, figsize=(13, 10))

# Display 'All data shared'
print('All data shared')
# Display the graph using the display_graph function and save the image
display(display_graph(ax, df_minerals_grid_final2, 'orig', ''))
filename = 'size_' + str(side_len) + '_alldatashared.png'

# Show the plot
plt.show()
```
